```python
import os
```

```python
prodir = "prompts_text_balanced"
imgdir = "preprocessed_fine_tune_imgs_balanced"
```

```python
if not os.path.exists(prodir):
    os.mkdir(prodir)
```

```python
views = os.listdir(imgdir)
```

```python
views
```

```python
base_prompts = "pixelated network data, "
```

```python
indi_prompts = {
    'attacktype': "attack type -%s",
    'cseg': "segment -%s",
    'proto': "protocol -%s",
    'time': "time %ss"
}
```

```python
for v in views:
    base_dir = os.path.join(imgdir, v)
    bpro_dir = os.path.join(prodir, v)
    if not os.path.exists(bpro_dir):
        os.mkdir(bpro_dir)
    
    cat_dict = {}
    code = 1
    
    imglist = os.listdir(base_dir)
    
    for i in imglist:
        cat = i.split(".png")[0].split("_")[1]
        if cat not in cat_dict.keys():
            if v == "time":
                cat_dict[cat] = cat
            else:
                cat_dict[cat] = code
                code = code + 1
        cur_code = cat_dict[cat]
        cur_pro = base_prompts + indi_prompts[v]%(cur_code)
        cur_proname = os.path.join(bpro_dir, f"{i.split('.png')[0]}.txt")
        print(cur_proname)
        with open(cur_proname, "w") as f:
            f.write(cur_pro)
```
